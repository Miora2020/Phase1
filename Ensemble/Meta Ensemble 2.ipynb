```python
# Importing packages
import pandas as pd
import numpy as np
from sklearn.cross_validation import train_test_split
from sklearn.cross_validation import KFold, cross_val_score
from sklearn.metrics import confusion_matrix,precision_recall_curve,
f1_score,auc,roc_auc_score,roc_curve,recall_score,classification_report,accuracy_score
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings("ignore")
from keras.models import Model, load_model
from sklearn.ensemble import RandomForestClassifier

"""Code for Credit Card Fraud Data"""

# Loading predictions of base classifiers on train and test data 
NN = pd.read_csv("tsmoteNN2.csv")
NNt = pd.read_csv("smoteNN2.csv")
XG = pd.read_csv("tsmoteXG2.csv")
XGt = pd.read_csv("smoteXG2.csv")
RF =pd.read_csv("tAErf2.csv")
RFt =pd.read_csv("AErf2.csv")
IF =pd.read_csv("tAEif2.csv")
IFt =pd.read_csv("AEif2.csv")
train = pd.read_csv('train2.csv')
test = pd.read_csv('test2.csv')
X =train.iloc[:,:-1]
Y = train.iloc[:,-1]
Yt = test.iloc[:,-1]

# Making a dataframe with columns representing predictions of each base classifier
A = pd.concat([NN,XG,RF,IF],axis=1)
A.columns=['NN','XG','RF','IF']
testX = pd.concat([NNt,XGt,RFt,IFt],axis=1)
testX.columns=['NN','XG','RF','IF']

# Hyper-parameter tuning og Random Forest-2
rf = RandomForestClassifier(random_state=1)
param_grid = {'n_estimators': range(1,51),'min_samples_split':range(2,51),
              'max_depth':range(1,51)}
CV_lr = GridSearchCV(estimator=rf,param_grid=param_grid,cv=5,scoring='f1',n_jobs=-1)
CV_lr.fit(X=A,y=Y)
best_param = CV_lr.best_params_
print("Best Paramters : ",best_param)

# Training performance
clf = RandomForestClassifier(n_estimators=1,min_samples_split=9,max_depth=5,
                             random_state=41,n_jobs=-1)
clf.fit(A, Y)
pred = clf.predict(A)
print(accuracy_score(Y,pred))
print(confusion_matrix(Y,pred))
print(classification_report(Y,pred))

# Test Set Predictions
pred = clf.predict(testX)
print(accuracy_score(Yt,pred))
print(confusion_matrix(Yt,pred))
print(classification_report(Yt,pred))

"""Code for the Abalone data"""

# loading predictions of each base clasifier
NN = pd.read_csv("tsmoteNN2.csv")
NNt = pd.read_csv("smoteNN2.csv")
XG = pd.read_csv("tsmoteXG2.csv")
XGt = pd.read_csv("smoteXG2.csv")
RF =pd.read_csv("tAErf2.csv")
RFt =pd.read_csv("AErf2.csv")
IF =pd.read_csv("tAEif2.csv")
IFt =pd.read_csv("AEif2.csv")
train = pd.read_csv('train2.csv')
test = pd.read_csv('test2.csv')
X =train.iloc[:,:-1]
Y = train.iloc[:,-1]
Yt = test.iloc[:,-1]
A = pd.concat([NN,XG,RF,IF],axis=1)
A.columns=['NN','XG','RF','IF']
testX = pd.concat([NNt,XGt,RFt,IFt],axis=1)
testX.columns=['NN','XG','RF','IF']

# Hyper-parameter tuning
rf = RandomForestClassifier(random_state=1,class_weight='balanced')
param_grid = {'n_estimators': range(1,11),'min_samples_split':range(2,11),
              'max_depth':range(1,11)}
CV_lr = GridSearchCV(estimator=rf,param_grid=param_grid,cv=5,scoring='f1',n_jobs=-1)
CV_lr.fit(X=A,y=Y)
best_param = CV_lr.best_params_
print("Best Paramters : ",best_param)

# Training results
clf = RandomForestClassifier(n_estimators=7,min_samples_split=2,max_depth=7,
                             random_state=41,class_weight='balanced',n_jobs=-1)
clf.fit(A, Y)
pred = clf.predict(A)
print(accuracy_score(Y,pred))
print(confusion_matrix(Y,pred))
print(classification_report(Y,pred))

# Test Set Predictions
pred = clf.predict(testX)
print(accuracy_score(Yt,pred))
print(confusion_matrix(Yt,pred))
print(classification_report(Yt,pred))
```
