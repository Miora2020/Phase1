```python
# Importing required packages
from keras.models import Sequential
from keras.layers import Dense
import numpy
# fix random seed for reproducibility
numpy.random.seed(7)
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix,precision_recall_curve,accuracy_score
from sklearn.metrics import auc,f1_score,roc_auc_score,roc_curve,recall_score
from sklearn.metrics import precision_score,recall_score,classification_report
import tensorflow as tf
from sklearn.model_selection import train_test_split
from keras.models import Model, load_model
from keras.layers import Input, Dense
from keras.callbacks import ModelCheckpoint, TensorBoard
from sklearn.preprocessing import StandardScaler
from keras import regularizers
from imblearn.over_sampling import SMOTE

# Importing data and preprocessing
data = pd.read_csv("creditcard.csv")
data['normAmount']=
        StandardScaler().fit_transform(data['Amount'].values.reshape(-1, 1))
data = data.drop(['Time','Amount'],axis=1)
X = data.iloc[:,data.columns != 'Class']
Y = data.iloc[:,data.columns == 'Class']
X_train,X_test,Y_train,Y_test=
            train_test_split(X,Y,test_size=0.20,random_state=21,stratify=Y)

# create model
model = Sequential()
model.add(Dense(21, input_dim=29, activation='relu'))
model.add(Dense(21, activation='relu'))
model.add(Dense(15, activation='tanh'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile model
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

# Fit the model
model.fit(X_train, Y_train, epochs=100, batch_size=256)

# evaluate the model
scores = model.evaluate(X_test, Y_test)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))

# calculate predictions
predictions = model.predict(X_test)
predictionst = model.predict(X_train)
# round predictions
rounded = [round(x[0]) for x in predictions]

# Training set reuslts for without undersampling case
rounded2 = [round(x[0]) for x in predictionst]
print("Train roc auc score is : ", roc_auc_score(Y_train,rounded2))
print("Accuracy : ",accuracy_score(Y_train,rounded2))
print("Classification report is : ")
print(classification_report(Y_train, rounded2))
print("Confusion Matrix for test set is : ")
print(confusion_matrix(Y_train,rounded2))
# Test set reuslts for without undersampling case
print("Test roc auc score is : ", roc_auc_score(Y_test,rounded))
print("Accuracy : ",accuracy_score(Y_test,rounded))
print("Classification report is : ")
print(classification_report(Y_test, rounded))
print("Confusion Matrix for test set is : ")
print(confusion_matrix(Y_test,rounded))

# Data splitting for undersampling
train= pd.concat([X_train, Y_train],axis=1) 
fraud = train[train["Class"]==1]
valid = train[train["Class"]==0]
# 10/90 Undersampling
# Model for 10/90 undersampling
valid90 = valid.sample(n=(394*10))
train90 = pd.concat([fraud, valid90])
train90 = train90.sample(frac=1).reset_index(drop=True)
X_train90 = train90.iloc[:,train90.columns!="Class"]
Y_train90 = train90.iloc[:,train90.columns=="Class"]
# create model
model = Sequential()
model.add(Dense(21, input_dim=29, activation='relu'))
model.add(Dense(21, activation='relu'))
model.add(Dense(15, activation='tanh'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile model
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

# Fit the model
model.fit(X_train90, Y_train90,epochs=100, batch_size=256)

# evaluate the model
scores = model.evaluate(X_test, Y_test)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))

# calculate predictions
predictions = model.predict(X_test)
predictionst = model.predict(X_train90)
# round predictions
rounded = [round(x[0]) for x in predictions]

# Training performance for 10/90 undersampling
zrounded2 = [round(x[0]) for x in predictionst]
print("Train roc auc score is : ", roc_auc_score(Y_train90,rounded2))
print("Accuracy : ",accuracy_score(Y_train90,rounded2))
print("Classification report is : ")
print(classification_report(Y_train90, rounded2))
print("Confusion Matrix for test set is : ")
print(confusion_matrix(Y_train90,rounded2))
# Test performance for 10/90 undersampling
print("Test roc auc score is : ", roc_auc_score(Y_test,rounded))
print("Accuracy : ",accuracy_score(Y_test,rounded))
print("Classification report is : ")
print(classification_report(Y_test, rounded))
print("Confusion Matrix for test set is : ")
print(confusion_matrix(Y_test,rounded))

# 20/80 Undersampling
valid80 = valid.sample(n=(394*4))
train80 = pd.concat([fraud, valid80])
train80 = train80.sample(frac=1).reset_index(drop=True)
X_train80 = train80.iloc[:,train80.columns!="Class"]
Y_train80 = train80.iloc[:,train80.columns=="Class"]
# Training
# create model
model = Sequential()
model.add(Dense(21, input_dim=29, activation='relu'))
model.add(Dense(21, activation='relu'))
model.add(Dense(15, activation='tanh'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile model
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

# Fit the model
model.fit(X_train80, Y_train80, epochs=100, batch_size=256)

# evaluate the model
scores = model.evaluate(X_test, Y_test)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))

# calculate predictions
predictions = model.predict(X_test)
predictionst = model.predict(X_train80)
# round predictions
rounded = [round(x[0]) for x in predictions]

# Training and Test Performance
rounded2 = [round(x[0]) for x in predictionst]
print("Train roc auc score is : ", roc_auc_score(Y_train80,rounded2))
print("Accuracy : ",accuracy_score(Y_train80,rounded2))
print("Classification report is : ")
print(classification_report(Y_train80, rounded2))
print("Confusion Matrix for test set is : ")
print(confusion_matrix(Y_train80,rounded2))
print("Test roc auc score is : ", roc_auc_score(Y_test,rounded))
print("Accuracy : ",accuracy_score(Y_test,rounded))
print("Classification report is : ")
print(classification_report(Y_test, rounded))
print("Confusion Matrix for test set is : ")
print(confusion_matrix(Y_test,rounded))

# 50/50 Undersampling
valid50 = valid.sample(n=(394))
train50 = pd.concat([fraud, valid50])
train50 = train50.sample(frac=1).reset_index(drop=True)
X_train50 = train50.iloc[:,train50.columns!="Class"]
Y_train50 = train50.iloc[:,train50.columns=="Class"]

# create model
model = Sequential()
model.add(Dense(21, input_dim=29, activation='relu'))
model.add(Dense(21, activation='relu'))
model.add(Dense(15, activation='tanh'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile model
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

# Fit the model
model.fit(X_train50, Y_train50, epochs=100, batch_size=256)

# evaluate the model
scores = model.evaluate(X_test, Y_test)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))

# calculate predictions
predictions = model.predict(X_test)
predictionst = model.predict(X_train50)
# round predictions
rounded = [round(x[0]) for x in predictions]

# Training and test performance
rounded2 = [round(x[0]) for x in predictionst]
print("Train roc auc score is : ", roc_auc_score(Y_train50,rounded2))
print("Accuracy : ",accuracy_score(Y_train50,rounded2))
print("Classification report is : ")
print(classification_report(Y_train50, rounded2))
print("Confusion Matrix for test set is : ")
print(confusion_matrix(Y_train50,rounded2))
print("Test roc auc score is : ", roc_auc_score(Y_test,rounded))
print("Accuracy : ",accuracy_score(Y_test,rounded))
print("Classification report is : ")
print(classification_report(Y_test, rounded))
print("Confusion Matrix for test set is : ")
print(confusion_matrix(Y_test,rounded))

# 90/10 undersampling
valid10 = valid.sample(n=(44))
train10 = pd.concat([fraud, valid10])
train10 = train10.sample(frac=1).reset_index(drop=True)
X_train10 = train10.iloc[:,train10.columns!="Class"]
Y_train10 = train10.iloc[:,train10.columns=="Class"]
# create model
model = Sequential()
model.add(Dense(21, input_dim=29, activation='relu'))
model.add(Dense(21, activation='relu'))
model.add(Dense(15, activation='tanh'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile model
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

# Fit the model
model.fit(X_train10, Y_train10, epochs=100, batch_size=256)

# evaluate the model
scores = model.evaluate(X_test, Y_test)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))

# calculate predictions
predictions = model.predict(X_test)
predictionst = model.predict(X_train50)
# round predictions
rounded = [round(x[0]) for x in predictions]

# Training and test performance
rounded2 = [round(x[0]) for x in predictionst]
print("Train roc auc score is : ", roc_auc_score(Y_train50,rounded2))
print("Accuracy : ",accuracy_score(Y_train50,rounded2))
print("Classification report is : ")
print(classification_report(Y_train50, rounded2))
print("Confusion Matrix for test set is : ")
print(confusion_matrix(Y_train50,rounded2))
print("Test roc auc score is : ", roc_auc_score(Y_test,rounded))
print("Accuracy : ",accuracy_score(Y_test,rounded))
print("Classification report is : ")
print(classification_report(Y_test, rounded))
print("Confusion Matrix for test set is : ")
print(confusion_matrix(Y_test,rounded))

# Importing data for oversampling (SMOTE Borderline-2)
train = pd.read_csv('train.csv')
weight = pd.read_csv('weight.csv')
test = pd.read_csv('test.csv')
weightX = weight.iloc[:,:-1]
weightY = weight.iloc[:,-1]
testX = test.iloc[:,:-1]
testY = test.iloc[:,-1]
X = train.iloc[:,:-1]
Y = train.iloc[:,-1]
sm= SMOTE(kind = 'borderline2')
X_resampled, Y_resampled = sm.fit_sample(X, Y)
# create model
model = Sequential()
model.add(Dense(21, input_dim=29, activation='relu'))
model.add(Dense(21, activation='relu'))
model.add(Dense(15, activation='tanh'))
model.add(Dense(10, activation='relu'))
model.add(Dense(6, activation='tanh'))
model.add(Dense(1, activation='sigmoid'))
# Compile model
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
# Fit the model
model.fit(X_resampled, Y_resampled, epochs=150, batch_size=512)
# calculate predictions
predictions = model.predict(X_test)
# round predictions
rounded = [round(x[0]) for x in predictions]
# Training test and weight tuninig set results
pred4 = model.predict(X_resampled)
pred4 = [round(x[0]) for x in pred4]
print("Test roc auc score is : ", roc_auc_score(Y_resampled,pred4))
print("Classification report is : ")
print(classification_report(Y_resampled,pred4))
print("Confusion Matrix for weight set is : ")
print(confusion_matrix(Y_resampled,pred4))
pred2 = model.predict(weightX)
pred2 = [round(x[0]) for x in pred2]
print("Test roc auc score is : ", roc_auc_score(weightY,pred2))
print("Classification report is : ")
print(classification_report(weightY,pred2))
print("Confusion Matrix for weight set is : ")
print(confusion_matrix(weightY,pred2))
pd.DataFrame(pred2, columns=['Class']).to_csv('WsmoteNN.csv',index = False)
pred3 = model.predict(testX)
pred3 = [round(x[0]) for x in pred3]
print("Test roc auc score is : ", roc_auc_score(testY,pred3))
print("Classification report is : ")
print(classification_report(testY,pred3))
print("Confusion Matrix for weight set is : ")
print(confusion_matrix(testY,pred3))
pd.DataFrame(pred3, columns=['Class']).to_csv('smoteNN.csv',index = False)
```
